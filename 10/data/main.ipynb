```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
df = pd.read_csv('population.csv')
df.head()
```

```python
# 転入・転出・計をplot
fig = plt.figure(figsize=(10, 5))
plt.plot(df['年度'], df['計'], label='計')
plt.legend()
plt.xticks(rotation=45)
plt.show()

# 転入・転出・計をplot
fig = plt.figure(figsize=(10, 5))
plt.plot(df['年度'], df['転入'], label='転入')
plt.plot(df['年度'], df['転出'], label='転出')
# plt.plot(df['年度'], df['計'], label='計')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
